```python
import numpy as np
import numpy.linalg as lin

a = np.array([[1.0,3.0],[5.0,7.0],[11.0,13.0]])
print(a)
print(a.shape)

#정방행렬이 아니면 pinv, 정방행렬이라면 inv (역행렬)
inv_a = lin.pinv(a)
print(inv_a)

b = np.array([17.0,19.0,23.0]).reshape(3,1)
print(b)
print(b.shape)

c = np.matmul(inv_a, b)
print(c)  

prey = c[0,0] * 1 + c[1,0] * 3
print(prey)

prey = c[0,0] * 5 + c[1,0] * 7
print(prey)

prey = c[0,0] * 11 + c[1,0] * 13
print(prey)

x1=-7.75
x2=8.25
```
